```python
def parse_v(vtxt):
    """ given text of the form 
    200215/isis/h_diagnostics/SiXIV/SiXIV_output.txt:V(K1) = -556.461 (-605.824,-509.619)
    (as produced by /Users/corcoran/research/ETA_CAR/CHANDRA2/repro/seqid/get_sixiv_vels.csh)
    parses each line and returns the velocity along with plus and minus error bars
    """
    seq=vtxt[0:6]
    b=vtxt.split("=")
    c=b[1].split("(")
    v=float(c[0])
    d=c[1].split(',')
    minus_error=v-float(d[0])
    plus_error=float(d[1].split(')')[0])-v
    return seq, v, minus_error, plus_error

def plot_echetgs_velocities(dummy,ion="sixiv",ylim=[-800,200],vdir="/Users/corcoran/research/ETA_CAR/CHANDRA2/repro/seqid/", 
                            obsinfo="/Users/corcoran/research/ETA_CAR/CHANDRA2/Observation_Info.txt", loc=8):
    import numpy as np
    fname=vdir+"/get_"+ion+"_vels.out"
    print "opening "+fname
    datafile=open(vdir+"/get_"+ion+"_vels.out","r")
    vel=datafile.read()
    datafile.close()
    b=vel.split("\n")
    v=[] 
    merror=[] # minus error
    perror=[] # plus error
    seq=[]
    for i in arange(len(b)-1):
        s,v1,m1,p1=parse_v(b[i])
        seq.append(s)
        v.append(v1)
        merror.append(m1)
        perror.append(p1)
    v=np.asarray(v)
    merror=np.asarray(merror)
    perror=np.asarray(perror)
    #
    # get obsinfo
    #
    obsinfofile=open(obsinfo,"r")
    obs=obsinfofile.read()
    obsinfofile.close()
    b=obs.split("\n")
    cols=b[0]
    b=b[1:] # strip off header
    oseq=[] 
    ojd=[]
    ophi=[]
    otime=[]
    for i in range(len(b)-1):
        c=b[i].split()
        oseq.append(c[0])
        ojd.append(float(c[1]))
        ophi.append(float(c[2]))
        otime.append(c[3])
    jd=[]
    phi=[]
    for s in seq:
        try:
            i=oseq.index(s)
            jd.append(ojd[i])
            phi.append(ophi[i])
        except ValueError:
            print s+" NOT IN OSEQ"
            jd.append(-99)
            phi.append(-99)
    phi=np.asarray(phi)
    jd=np.asarray(jd)
    #
    # indices for cycles 1, 2, 3 (cycles centered on the minima)
    #
    cyc0=where((floor(phi-0.5))==0)
    cyc0=cyc0[0]
    cyc1=where(floor(phi-0.5)==1)
    cyc1=cyc1[0]
    cyc2=where(floor(phi-0.5)==2)
    cyc2=cyc2[0]
    #
    # plot
    #
    fig=figure(figsize=(8,5))
    #axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    x=phi-floor(phi)
    """
    axes.set_ylim(ylim)
    axes.set_xlim([-0.1,1.1])
    axes.errorbar(x, v, yerr=[merror, perror], fmt='o', label="Cycle 0") # labels are what appear in the legend
    axes.plot(x[cyc1], np.asarray(v)[cyc1], 'ro', label="Cycle 1") # need to convert v from a list to a numpy array
    axes.plot(x[cyc2], np.asarray(v)[cyc2], 'wo', label="Cycle 2")
    axes.set_title('Velocities for '+ion.upper(), fontsize=18)
    axes.set_xlabel('Phase', fontsize=18)
    axes.set_ylabel('Centroid Velocity (km/s)', fontsize=18)
    axes.legend(scatterpoints=1,loc=8) # scatterpoints =1 is supposed to generate a legend with only 1 scatterpoint symbol
    """
    xlim([-0.1, 1.1])
    xlabel("Phase", fontsize=18)
    ylabel("Velocity", fontsize=18)
    title(ion, fontsize=18)
    scatter(x[cyc0], np.asarray(v)[cyc0], marker="o", label="Cycle 0", color="b")
    errorbar(x[cyc0], np.asarray(v)[cyc0], yerr=[merror[cyc0], perror[cyc0]], color="b", linestyle="none")
    scatter(x[cyc1], np.asarray(v)[cyc1], marker="o", label="Cycle 1", color="r")
    errorbar(x[cyc1], np.asarray(v)[cyc1], yerr=[merror[cyc1], perror[cyc1]], color="r", linestyle="none")
    scatter(x[cyc2], np.asarray(v)[cyc2], marker="o", label="Cycle 2", color="g")
    errorbar(x[cyc2], np.asarray(v)[cyc2], yerr=[merror[cyc2], perror[cyc2]], color="g", linestyle="none")
    legend(scatterpoints=1, loc=loc)

    return seq,jd,phi,v,oseq,ojd,ophi,otime
```
