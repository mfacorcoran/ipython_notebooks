We want to find all the table rows and reverse the ordering of the rows.  A row starts with <TR and ends with /TR>.  So if we find a line at index i in the array lines which has the <TR character, that means that lines[i] is the start of a row.  Once we find that line, we can find the next occurrence of /TR> (at index k, say), then the given row spans line[i:k].

So use this algorithm

-  Read the file in as a string (called "a")
-  while the length of a is >0, 
 -  use the "find" method to get the location in "a" of <TR> (if <TR> not found, set length of a to 0)
 -  extract the substring starting with <TR> and ending with </TR>
 -  save this as the row, then prepend it to the "rows" list
 -  remove this substring from "a" then repeat the procedure 



```python
ddir="/Users/corcoran/program/heapow/revised/revised_index_files/stars"
iname=ddir+"/index.html"
file=open(iname,"r")
a=file.read()
a=a.replace('\n',' ') # replace the \n with a blank space
a=a.replace('\t',' ') # replace the \t with a blank space
# get the html before the start of the rows
x=a.find("<TR>")
htmlstart=a[0:x-1]
rows=['']
while (len(a)>0):
    x=a.find("<TR>") # this finds the first occurrence of the start of a row
    if x>=0:
        y=a.find("</TR>") # this finds the first occurrence of the end of the row
        xstart=x
        yend=y+5
        row=a[xstart:yend] # this is the row - it starts at x and goes to y+5 (since </TR> has 5 characters
        print "row = %s\n" % row
        rows=[row]+rows
        a=a[yend+1:]
    else: 
        print "No more rows"
        htmlend=a[0:]
        rows=[htmlstart]+rows+[htmlend]
        a=''
```
