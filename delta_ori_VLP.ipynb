<h3><a name="f2ijoy">Reading & plotting $f/i$ ratios from Joy's excel file</a></h3>

```python
def calc_f2i(ion,xldir='/Users/corcoran/research/Delta_Ori/VLP/Data/ao14/work/he_diagnostics_joy/20130916/line_statistics', chat=0):
    """
    for the specified ion, extracts the f/i ratio from Joy's spreadsheets
    returns the f/i ratio and errors, along with the times and phases 
    corresponding to each time slice
    
    ion should be either S15, Si13,Mg11, or O7
    set chat>0 to print diagnostics
    """
    import numpy as np
    from xlrd import open_workbook
    slicenum=range(1,13) # for each slice, 1 to 12
    #xldir='/Users/corcoran/research/Delta_Ori/VLP/Data/ao14/work/he_diagnostics_joy/nichols_j'
    f2i=[]; f2iminus=[]; f2iplus=[]; slicetime=[] ; slicephase=[]
    sumfile= xldir+"/f2i_summary_mfc.xlsx"
    #
    # open summary table to get mjds and phases for each slice
    #
    if chat>0:
        print sumfile
    wbsum = open_workbook(sumfile)
    sheetsum=wbsum.sheet_by_index(0)
    colsum=sheetsum.col_values(0) # get values of column 0
    for s in slicenum:
            sindex=colsum.index(float(s))
            rowsum=sheetsum.row_values(sindex)
            slicetime.append((rowsum[2]+rowsum[4])/2.0) # add mjd midpoint for slixe
            slicephase.append(rowsum[7]) # col 7 contains the phase midpoint
            if chat>0:
                print "%i, %f, %f, %f, %f" % (s, rowsum[2],rowsum[4], (rowsum[2]+rowsum[4])/2.0, rowsum[7])
            #
            # now retrieve the fluxes, calc. f2i and +,- errors
            #
            xlname=xldir+"/"+"slice"+str(s)+".xlsx"
            wb = open_workbook(xlname)
            sheet=wb.sheet_by_index(0) # get first sheet in workbook
            col=sheet.col
            xlname=xldir+"/"+"slice"+str(s)+".xlsx"
            if chat>0:
                print "File is = %s" % xlname
            #ion=["S15","Si13", "Mg11","Ne9","O7"] 
            # Joy puts the fluxes in column 35 (counting from 0) with the errors in 36, 37 
            for i in range(len(col(0))):
                if col(0)[i].value==ion+"-i":
                    #print "Index is ",i
                    fluxi=col(35)[i].value # flux of the i-line
                    fluxerri=[col(36)[i].value, col(37)[i].value]
                    fluxf=col(35)[i+1].value # flux of the i-line
                    fluxerrf=[col(36)[i+1].value, col(37)[i+1].value]
                    if chat>0:
                        print "ion is %s, fluxi is %e (errmin=%e, errplus=%e)" % (ion, fluxi, fluxerri[0], fluxerri[1])
                    if chat>0:
                        print "ion is %s, fluxf is %e (errmin=%e, errplus=%e)" % (ion, fluxf, fluxerrf[0], fluxerrf[1])
                    difimin=fluxi-fluxerri[0]
                    if difimin<0:
                        difimin=0
                    diffmin=fluxf-fluxerrf[0]
                    if diffmin<0:
                        diffmin=0
                    try:
                        f2imax=(fluxf+fluxerrf[1])/difimin
                    except ZeroDivisionError:
                        try:
                            f2imax=(fluxf+fluxerrf[1])/fluxi
                        except ZeroDivisionError:
                            f2imax=float('Nan')
                    try:
                        f2imin=diffmin/(fluxi+fluxerri[1])
                    except ZeroDivisionError:
                        try:
                            f2imin=diffmin/fluxi
                        except ZeroDivisionError:
                            f2imin=float('Nan')
                    try:
                        f_2_i=fluxf/fluxi
                    except ZeroDivisionError:
                        f_2_i=float('Nan')
                    f2i.append(f_2_i)
                    f2iminus.append(f_2_i-f2imin)
                    f2iplus.append(f2imax-f_2_i)
                    if chat> 0:
                        print "f2i = %f (-%f, +%f), f2imax = %f" % (f_2_i, f_2_i-f2imin, f2imax-f_2_i, f2imax)
    return np.asarray(f2i), np.asarray(f2iminus), np.asarray(f2iplus), np.asarray(slicetime), np.asarray(slicephase)
            
    
```

```python
def readaglc_ion(ion, tzero=472320000, 
                 fdir="/Users/corcoran/Dropbox/DELTA_ORI/Analysis_Results/x-ray_photometry/order123/lines/"):
    """
    This routine reads FITS files output from the isis scripts in 
    /Users/corcoran/research/Delta_Ori/VLP/Data/ao14/work/isis_analysis/from_joy/lc_analysis
    These scripts create dispersed lightcurves for the delta ori spectra for different time bins and wavelength intervals
    ion should be one of 
       SiXIII
       NeX
       MgXI
       OVIII
       FeXVII
    """
    import pyfits
    import numpy as np
    import glob
    lc=glob.glob(fdir+"/"+ion+"/*.lc")
    lc_bg=glob.glob(fdir+"/"+ion+"/*.bkg")
    time_tot=np.asarray([])
    rate_tot=np.asarray([])
    rate_err_tot=np.asarray([])
    for i in [0,1,2,3]:
        #print lc[i]
        hdulist = pyfits.open(lc[i])
        lchd=hdulist[1].header
        lcdata=hdulist[1].data
        hdulistbg=pyfits.open(lc_bg[i])
        bgdata=hdulistbg[1].data
        counts_bg=bgdata.field("counts")
        cols=hdulist[1].columns
        time=lcdata.field("time")+lchd['TIMEZERO']
        counts=lcdata.field("counts")
        counts_err=lcdata.field("stat_err")
        expo_bin=lcdata.field("exposure")
        counts_bg_err=bgdata[1].field("stat_err")
        rate=(counts-counts_bg)/expo_bin
        #print counts, counts_bg, expo_bin
        rate_err=sqrt((counts_err/expo_bin)**2+(counts_bg_err/expo_bin)**2)
        j=where(np.isnan(expo_bin)==False) # only accept those values with non-Null exposures
        time_tot=np.append(time_tot,time[j])
        rate_tot=np.append(rate_tot,rate[j])
        rate_err_tot=np.append(rate_err_tot,rate_err[j])
    isort=np.argsort(time_tot)
    time_tot=time_tot[isort]
    rate_tot=rate_tot[isort]
    rate_err_tot=rate_err_tot[isort]
    fig=plt.figure(figsize=(12, 6))
    return time_tot, rate_tot, rate_err_tot

tzero=472320000

fdir="/Users/corcoran/Dropbox/DELTA_ORI/Analysis_Results/x-ray_photometry/order123/lines/"
ion="SiXIII"
time_totsi13, rate_totsi13, rate_err_totsi13 = readaglc_ion(ion, fdir=fdir)
xlabel("SC time - "+str(tzero/86400)+" (days)",fontsize=18)
ylabel("Net Count Rate",fontsize=18)
title(" Binsize = 15 ksec, ion= "+ion, fontsize=12)
axis(ymin=0.0,ymax=.02)
errorbar((time_totsi13-tzero)/86400.,rate_totsi13,rate_err_totsi13,fmt="ro",capsize=0, label=ion)
legend(loc=2, scatterpoints = 1)
```
